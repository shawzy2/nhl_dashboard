## Roster Data
scrape

```python
# list of teamIds
SEASON_ID = '20232024'
team_tricodes = sorted(df_teams['triCode'].tolist())
new_roster = []
for tricode in team_tricodes:
    # download roster
    r = requests.get(url=f'https://api-web.nhle.com/v1/roster/{tricode}/{SEASON_ID}')
    d = r.json()
    
    # extract all players
    roster = []
    for position, players in d.items():
        for player in players:
            roster.append(player)
    
    # download info about each player in roster
    for player in roster:
        try:
            # some rookies don't have jersey numbers or positions assigned
            jersey_num = 0
            if 'sweaterNumber' in player:
                jersey_num = player['sweaterNumber']
                
            # some players don't have birth cities
            birth_city = ''
            if 'birthCity' in player:
                birth_city = player['birthCity']['default']
                
            new_roster.append({
                'playerId': player['id'],
                'teamTricode': tricode,
                'birthDate': player['birthDate'],
                'firstName': player['firstName']['default'],
                'lastName': player['lastName']['default'],
                'primaryNumber': jersey_num,
                'birthCity': birth_city,
                'birthCountry': player['birthCountry'],
                'heightInInches': player['heightInInches'],
                'weightInPounds': player['weightInPounds'],
                'shootsCatches': player['shootsCatches'],
                'position': player['positionCode'],
                'headshotUrl': player['headshot']
            })
            
        except:
            print(f'error on playerId {player["id"]}, tricode {tricode}')
        
```

```python
df_roster = pd.DataFrame(new_roster)
df_roster.head()
```

```python
# merge teamId
df_roster = df_roster\
                .merge(df_teams[['triCode', 'teamId']], left_on='teamTricode', right_on='triCode')\
                .drop('triCode', axis=1)
df_roster
```

```python
# save roster data
df_roster.to_json('../../backend/data/rosterCurrent.json', orient='records')
```

## Headshot Images

```python
import time
```

```python
count = 0
for row in df_roster[['playerId', 'headshotUrl']].iterrows():
    # sleep so the api doesn't rate limit us
    if count % 20 == 0:
        print(f'through {count} players, sleeping 1 second')
        time.sleep(1)
    count += 1
    
    playerId = row[1]['playerId']
    img = requests.get(row[1]['headshotUrl']).content
    with open(f'../../frontend/src/assets/headshots/{playerId}.jpg', 'wb') as handler:
        handler.write(img)
```
